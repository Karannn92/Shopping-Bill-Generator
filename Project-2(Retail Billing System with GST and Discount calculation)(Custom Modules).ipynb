```python
#Retail Billing System with GST and Discount Calculation.
with open('bills.py','w') as f:
    f.write("""
def calc_total(price, quantity):
    return price * quantity
def calc_gst(total, gst_rate = 0.18):
    return total * gst_rate
def calc_discount(amt):
    if amt > 1000 : 
        discount_amt = amt * 0.10
        return True, discount_amt
    else:
        return False, 0
""")
```

```python
import bills
```

```python
# input
product_name = input("Enter the Product :")
price = float(input("Enter price of product: "))
quantity = int(input("Enter quantity: "))

total = bills.calc_total(price, quantity)
gst = bills.calc_gst(total)
final_amt = total + gst

discount, discount_amt = bills.calc_discount(final_amt)

final_amt = final_amt - discount_amt

# Display Bill
print("\n========== SHOPPING BILL ==========")
print("Product Name      :", product_name)
print("Price per item    :", price)
print("Quantity          :", quantity)
print("Total Amount      :", total)
print("GST (18%)         :", gst)
print("Discount Applied  :", discount)
print("Discount Amount   :", discount_amt)
print("Final Bill Amount :", final_amt)
print("==================================")
```
